```python
import numpy as np
import re
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from numpy import savetxt
from scipy.stats import zscore
```

```python
# Reading data

alphabet_data = pd.read_excel("E:/Uni/ML/HW1/HW1_AUT_MLPR_4021-Ar-Fa.xlsx")
alphabet_data.head()
```

```python
# Adding index column

index_col = list()
for i in range(alphabet_data.shape[0]):
    index_col.append(f"sentence {i+1}")
alphabet_data.insert(0,'ID', index_col)
alphabet_data = alphabet_data.set_index('ID')
alphabet_data.head()
```

```python
names = alphabet_data.index
alphabet_data.shape
```

```python
# Breaking setences to characters

bow = list()
unique_char = set()

for i in range(alphabet_data.shape[0]):
    dic = dict()
    text_without_whitespace = re.sub(r'\s+', '', alphabet_data.iloc[i, 0])
    for char in text_without_whitespace:
        if char not in dic:
            unique_char.add(char)
            dic[char] = 0
        dic[char] += 1
    bow.append(dic)

# Printing

for row in bow:
    print(row)
```

```python
df = pd.DataFrame(bow)
df = df.fillna(0)
df = df.astype(int)
df.head()
```
